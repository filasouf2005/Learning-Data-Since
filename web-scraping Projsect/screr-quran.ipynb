```python
import os
import requests
import time

desktop = os.path.join(os.path.expanduser("~"), "Desktop")
SAVE_FOLDER = os.path.join(desktop, "القرآن الكريم")
os.makedirs(SAVE_FOLDER, exist_ok=True)

BASE_URL = "https://ia902802.us.archive.org/28/items/www.elahmad.com_69/"

for i in range(3, 115):
    file_number = f"{i:03d}.mp3"
    mp3_url = BASE_URL + file_number
    file_path = os.path.join(SAVE_FOLDER, file_number)

    print(f"Downloading: {mp3_url}")

    success = False

    for attempt in range(1, 6):
        try:
            response = requests.get(mp3_url, timeout=30)  # ⬅ زيادة الوقت
            if response.status_code == 200:
                with open(file_path, "wb") as f:
                    f.write(response.content)
                print(f"✔ Saved: {file_path}\n")
                success = True
                break
            else:
                print(f"❌ Not Found: {file_number} (Status {response.status_code})")
                break

        except requests.exceptions.Timeout:
            print(f"⏳ Timeout on attempt {attempt} for {file_number}, retrying...")
            time.sleep(3)  # انتظار 3 ثوانٍ وإعادة المحاولة

        except Exception as e:
            print(f"❌ Error on attempt {attempt}: {e}")
            time.sleep(3)

    if not success:
        print(f"❌ Failed to download {file_number} after 3 attempts.\n")
```
