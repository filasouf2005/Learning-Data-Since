# Create Index


```python
new_index=np.arange(1000)
#shuffling
np.random.shuffle(new_index)
new_index
```

```python
df_exems['new_index']=new_index
df_exems
```

```python
#Set an index
df_exems.set_index('new_index',inplace=True)
```

```python
#sort the index

df_exems.sort_index(inplace=True)
```

### Rename Columns


```python
#rename coulmn 
df_exems.rename(columns={'gender':'Gender'},inplace=True)
df_exems
#rename multiple columns
df_exems.rename(columns={'parental level of education':'PLE','test preparation course':'TPC','reading score':'RS','math score':'MC','writing score':'WC','language score':'LC'},inplace=True)
df_exems
```

```python
#rename the index 0,1....and update the dataFrame 
df_exems.rename(index={0:'A',1:'B'},inplace=True)
df_exems
```
