### np.select

```python
import numpy as np

#creat an array based on multiple price tiers
conditions = [
    df_laptop['Price_euros']>3000,
    (df_laptop['Price_euros']>2000) & (df_laptop['Price_euros']<=3000) ,
    (df_laptop['Price_euros']>800) & (df_laptop['Price_euros']<= 2000),
    (df_laptop['Price_euros']<=800)
]

values =  ['Too Expensive','Expensive','Affortable','Cheap']

# setit to a new column
df_laptop['Price_tiers'] = np.select(conditions,values, default='Unknown')
df_laptop

```

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small<12

# create "conditions" and "values" variables
conditions = [
    (df_laptop['Inches']>16),
    (df_laptop['Inches']>14) & (df_laptop['Inches']<=16),
    (df_laptop['Inches']>12) & (df_laptop['Inches']<=14),
    (df_laptop['Inches']<=12),
]
choices = ['Too Big', 'Big', 'Small', 'Too Small']
# set it to a new column
df_laptop['Screen_size'] = np.select(conditions, choices)
# show dataframe
df_laptop
```

## Duplicate


```python
# find duplicates in 1 column/series
df_laptop.duplicated('laptop_ID')
# show elements in dataframe with duplicates in column 'laptop_ID'
df_laptop[df_laptop.duplicated('laptop_ID')]
```

```python
# duplicates in 2 or more columns
duplicated = df_laptop.duplicated(['Product', 'TypeName', 'Inches'])
# show all the values duplicated (except for the first - keep='first')
df_laptop[duplicated].sort_values(['Product', 'TypeName'])
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops = df_laptop.sort_values(['Company', 'Price_euros'])
df_laptops
# check all value categories in "Company" column
df_laptops.value_counts('Company')
# duplicated values in "Company" columns (default -> keep='first')
duplicated_first = df_laptops.duplicated('Company')
# show dataframe with duplicated values in "Company" column
df_laptops[duplicated_first]

# keep='first'(cheapest laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptops[~duplicated_first]
df_laptops[~duplicated_first][['Company', 'Price_euros']]

# check all value categories
# df_laptops[~duplicated].value_counts('Company')
```

```python
# keep='last'
duplicated_last = df_laptops.duplicated('Company', keep='last')
# keep=False
duplicated_false = df_laptops.duplicated('Company', keep=False)
# keep='last'(most expensive laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptops[~duplicated_last]
df_laptops[~duplicated_last][['Company', 'Price_euros']]
```
