## Drop Rows OR Columns

```python
#Drop one row
df.drop('L. Messi',axis=0)

#Drop tow or more rows
df.drop(['L. Messi','Cristiano Ronaldo'],axis=0)

```

```python
#Drop one col
df.drop('long_name',axis=1)
#drop the last col
df.drop(df.columns[[-1]],axis=1)

#Drop tow or more cols
df.drop(['long_name','dob'],axis=1)

```

## Create a Random Sample

## Filter with query()

```python
#select player older than34 
df.query("age>=34")
```

```python
#select players older than 34 from Italy
df.query("age>=34 and nationality=='Italy' ")
```

```python
#Add not oprator to the first query
df.query("not(age>=34)")
```

```python
#convert height to meters and select those with height above 1.8
df.query("height_cm/100>1.8")
```

```python
#select players that were born after 1990

#check out data types
df.dtypes
#conver 'dob' col datetime type
df['dob'] = df['dob'].astype('datetime64[ns]')

#query

df.query('dob.dt.year >1990')
```

#### Aplly method


```python
#use numpy fun and apply it to series
import numpy as np
df['age'].apply(np.sqrt)
```

```python
#creat your own fun and apply it to  a data frame
def calc_bmi(row):
    return row['weight_kg']/(row['height_cm']/100)**2
df.apply(calc_bmi,axis=1)
```
